## Data Extract

```python
# outer box : <div class="css-5wh65g"
# name of product : <span class="_0T8-iGxMpV6NEsYEhwkqEg==">Seblak Instan Premium Bangsatria</span>
# price : <div class="_67d6E1xDKIzw+i2D2L0tjw== ">Rp21.000</div>
# seller : <span class="T0rpy-LEwYNQifsgB-3SQw== pC8DMVkBZGW7-egObcWMFQ== flip">Foodstocks</span>
# city : <span class="pC8DMVkBZGW7-egObcWMFQ== flip">Jakarta Selatan</span>
# sold : <span class="se8WAnkjbVXZNA8mT+Veuw==">100+ terjual</span>
# rating : <span class="_9jWGz3C-GX7Myq-32zWG9w==">5.0</span>
```

```python
#container
product = []
price   = []
seller  = []
city    = []
sold    = []
rating  = []

#running driver
driver = webdriver.Chrome() #initialize the Chrome webdriver
for page in range(1,15):    #define range of page will be scrapped
        driver.get(f'https://www.tokopedia.com/search?navsource=&page={page}&q=seblak%20instan&search_id=202501281605537D715104E2EFC915BQA6&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st=')  #Replace the page number in the URL template
        driver.maximize_window() # maximize the browser window
        scheight = .1 #initialize the scroll height
        #scroll through the page to load all content
        while scheight < 9.9: 
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight/%s);" % scheight)
                scheight += .005
        html = driver.page_source   #get the HTML source of the page
        soup = BeautifulSoup(html, 'html.parser') #parse the HTML using BeautifulSoup

#scrapping data
        for data in soup.find_all('div', {'class':'css-5wh65g'}): #outer box html
                try:
                        product.append(data.find('span', {'class':'_0T8-iGxMpV6NEsYEhwkqEg=='}).get_text()) #retrieve name of the product from the website
                except:
                        product.append(None) #fill 'None' if name of product is not attached
                try:
                        price.append(data.find('div', {'class':'_67d6E1xDKIzw+i2D2L0tjw=='}).get_text()) #retrieve data price from the website
                except:
                        price.append(None) 
                try:
                        seller.append(data.find('span', {'class':'T0rpy-LEwYNQifsgB-3SQw== pC8DMVkBZGW7-egObcWMFQ== flip'}).get_text()) #retrieve data seller from the website
                except:
                        seller.append(None) 
                try:
                        city.append(data.find('span', {'class':'pC8DMVkBZGW7-egObcWMFQ== flip'}).get_text()) #retrieve data city of the seller from the website
                except:
                        city.append(None) 
                try:
                        sold.append(data.find('span', {'class':'se8WAnkjbVXZNA8mT+Veuw=='}).get_text()) # retrieve data of number product sold from the website
                except:
                        sold.append(None) 
                try:
                        rating.append(data.find('span', {'class':'_9jWGz3C-GX7Myq-32zWG9w=='}).get_text()) #retrieve data rating from the website
                except:
                        rating.append(None)

driver.close()
```

```python
#Verify that all columns have the same number of rows
print(len(product))
print(len(price))
print(len(seller))
print(len(city))
print(len(sold))
print(len(rating))
```

```python
# Import Library Panda
import pandas as pd

#create DataFrame
df = pd.DataFrame()

df['Product'] = product
df['Price']   = price
df['Seller']  = seller
df['City']    = city
df['Rating']  = rating
df['Sold']    = sold
```

## Data Transform

### Exploratory Data Analysis

```python
df.info()
```

total rows is 1140

Data type consistency: from data type, we need to change price, rating and sold to be integer/float

Data incompleteness: The "Rating" column has 724 rows of missing data, while the "Sold" column has 400 rows of missing data. This indicates that the data is incomplete.

### Data Cleaning

```python
df['Price'] = df['Price'].str.replace('\D', '', regex=True).astype(int) #change datatype price to integer
```

```python
df['Rating'] = df['Rating'].astype(float) #change datatype price to integer
```

```python
df.Sold = df.Sold.str.replace('rb','000', regex=False) #change rb(thousand) to be 000
df['Sold'] = df['Sold'].str.replace('\D', '', regex=True)  #remove 'terjual'
df["Sold"]= df["Sold"].fillna(0).astype(int) #change datatype sold to integer
```

## Data Loading

```python
import psycopg2
from psycopg2 import sql

# PostgreSQL connection details
db_user = "postgres"
db_password = "postgres"
db_host = "localhost"
db_port = "5432"

# Create a connection to the PostgreSQL database
connection = psycopg2.connect(
    user=db_user,
    password=db_password,
    host=db_host,
    port=db_port,
    database="seblak_tokopedia"
)
```

```python
cursor = connection.cursor()
connection.autocommit = True

q = sql.SQL('''CREATE TABLE IF NOT EXISTS data_seblak_instan (
    id SERIAL PRIMARY KEY,
    product TEXT,
    price INTEGER,
    seller VARCHAR(100),
    city TEXT,
    rating FLOAT,
    sold INTEGER
);''')

cursor.execute(q)
```
